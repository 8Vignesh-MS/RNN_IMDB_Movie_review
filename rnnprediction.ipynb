```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
```

```python
#load imdb dataset word index
word_index=imdb.get_word_index()
reverse_word_index = {value:key for key,value in word_index.items()}
```

```python
model = load_model("C:\Users\8vign\Desktop\gen\simplernn\simple_rnn_imdb.h5")
model.summary()
```

```python
model.get_weights()
```

```python
#function to decode reviews
def decode_review(encoded_review):
  return " ".join([reverse_word_index.get(i-3,'?') for i in encoded_review])

  #function to preprocess user input
max_features = 1000
def preprocess_text(text):
    words = text.lower().split()
    encoded_review = []

    for word in words:
        index = word_index.get(word, 2)
        if index < max_features:     # ✅ restrict vocabulary
            encoded_review.append(index + 3)
        else:
            encoded_review.append(2)  # OOV token

    padded_review = sequence.pad_sequences([encoded_review], maxlen=500)
    return padded_review
```

```python
#prediction function
def predict_sentiment(review):
  preprocessed_input=preprocess_text(review)

  prediction=model.predict(preprocessed_input)

  sentiment = "Positive" if prediction[0][0] > 0.5 else "Negative"

  return sentiment,prediction
```

```python
#example review

eg_review = "This movie was fantastic! the acting was brilliant and the plot was thrilling"
sentiment,score = predict_sentiment(eg_review)

print(f"Review: {eg_review}")
print(f"Sentiment: {sentiment}")
print(f"Score: {score}")
```
